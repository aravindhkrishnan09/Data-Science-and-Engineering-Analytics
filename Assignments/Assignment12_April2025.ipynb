# Mapping

```python
#mapping
import pandas as pd

a = {'name': ['John', 'Jane', 'Jim', 'Jack'], 'profession': ['doctor', 'engineer', 'teacher', 'artist']}
df = pd.DataFrame(a)
df['profession_code'] = df['profession'].map({'doctor': 1, 'engineer': 2, 'teacher': 3, 'artist': 4})
print(df)
```

# Apply

```python
import numpy as np
import pandas as pd

a = np.random.randint(1, 100,5)
a_apply = pd.Series(a).apply(lambda x: x**2)
#using list comprehension
b_apply = pd.Series([x**2 for x in a])
print(a)
print(a_apply)
print(b_apply)
```

# Eval function

```python
#evaluate
import pandas as pd
import numpy as np
a = np.random.randint(1, 100, 5)
a_eval = pd.Series(a).apply(lambda x: eval('x**2'))
#using list comprehension
b_eval = pd.Series([eval('x**2') for x in a])
print(a)
```
