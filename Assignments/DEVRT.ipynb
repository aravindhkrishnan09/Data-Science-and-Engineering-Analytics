```python
import os
import pandas as pd

def load_files_from_directory(directory_path, file_types=('csv', 'xls', 'xlsx')):
    """
    Load and append CSV and Excel files from a given directory into a single pandas DataFrame.

    Args:
        directory_path (str): Path to the directory containing the files.
        file_types (tuple): File types to load. Default is ('csv', 'xls', 'xlsx').

    Returns:
        pd.DataFrame: A single DataFrame containing data from all files.
    """
    dataframes = []
    for filename in os.listdir(directory_path):
        filepath = os.path.join(directory_path, filename)
        if os.path.isfile(filepath):
            ext = filename.lower().split('.')[-1]
            try:
                if ext == 'csv' and 'csv' in file_types:
                    df = pd.read_csv(filepath)
                    dataframes.append(df)
                elif ext in ['xls', 'xlsx'] and ext in file_types:
                    df = pd.read_excel(filepath)
                    dataframes.append(df)
            except Exception as e:
                print(f"Failed to load {filename}: {e}")
    if dataframes:
        return pd.concat(dataframes, ignore_index=True)
    else:
        return pd.DataFrame()  # Return an empty DataFrame if no files are loaded
```

```python
data_dir = r"G:\DIYguru\Data-Science-and-Engineering-Analytics\Assignments\DACIA SPRING"
combined_data_d = load_files_from_directory(data_dir)

data_dir = r"G:\DIYguru\Data-Science-and-Engineering-Analytics\Assignments\NISSAN LEAF"
combined_data_n = load_files_from_directory(data_dir)

print(f"Combined DataFrame shape: {combined_data_d.shape}")
print(f"Combined DataFrame shape: {combined_data_n.shape}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(id6['timestamp_data_utc'], id6['soc'], label='SOC', color='blue')
plt.title('SOC over Time for Route ID 16')
plt.xlabel('Time')
plt.ylabel('SOC')
plt.legend()
plt.show()

```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data for regression
X = (id6['timestamp_data_utc'] - id6['timestamp_data_utc'].min()).dt.total_seconds().values.reshape(-1, 1)
y = id6['soc'].values

# Fit linear regression
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(id6['timestamp_data_utc'], y, label='SOC', color='blue')
plt.plot(id6['timestamp_data_utc'], y_pred, label='Linear Regression', color='red', linestyle='--')
plt.title('Linear Regression: SOC over Time for Route ID 16')
plt.xlabel('Time')
plt.ylabel('SOC')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Fit a polynomial regression (degree 2 for example)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_poly_pred = poly_model.predict(X_poly)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(id6['timestamp_data_utc'], y, label='SOC', color='blue')
plt.plot(id6['timestamp_data_utc'], y_poly_pred, label='Polynomial Regression (deg=2)', color='green', linestyle='--')
plt.title('Polynomial Regression: SOC over Time for Route ID 16')
plt.xlabel('Time')
plt.ylabel('SOC')
plt.legend()
plt.show()
```
