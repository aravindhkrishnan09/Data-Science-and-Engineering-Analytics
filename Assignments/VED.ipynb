```python
import os
import random
import pandas as pd
from glob import glob

# Set your folders here
dir1 = r"G:\DIYguru\Data-Science-and-Engineering-Analytics\Assignments\other_source_files\VED_DynamicData_Part1"
dir2 = r"G:\DIYguru\Data-Science-and-Engineering-Analytics\Assignments\other_source_files\VED_DynamicData_Part2"

# Set sampling fraction or number of rows
sample_fraction = 0.05  # 5% sample
sample_rows = None      # Or set sample_rows = 1000 to get 1000 rows from each file

# Combine all CSV file paths
csv_files = glob(os.path.join(dir1, '*.csv')) + glob(os.path.join(dir2, '*.csv'))

# Function to sample a CSV file without loading all rows
def sample_csv(filepath, fraction=None, n_rows=None):
    with open(filepath) as f:
        # Count total number of lines
        total_lines = sum(1 for line in f) - 1  # excluding header

    if n_rows:
        skip = sorted(random.sample(range(1, total_lines + 1), total_lines - n_rows))
    elif fraction:
        keep_rows = int(total_lines * fraction)
        skip = sorted(random.sample(range(1, total_lines + 1), total_lines - keep_rows))
    else:
        raise ValueError("Provide either 'fraction' or 'n_rows'")

    return pd.read_csv(filepath, skiprows=skip, low_memory=False)

# Read and concatenate sampled data
sampled_dfs = []

for file in csv_files:
    try:
        df_sample = sample_csv(file, fraction=sample_fraction, n_rows=sample_rows)
        sampled_dfs.append(df_sample)
    except Exception as e:
        print(f"Skipping file {file} due to error: {e}")

# Final combined sampled DataFrame
df_dynamic = pd.concat(sampled_dfs, ignore_index=True)

print(f"Final sampled dataset shape: {df_dynamic.shape}")
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Reference datetime for DayNum = 1
reference_date = datetime(2017, 11, 1)

# Convert DayNum to timestamp using vectorized timedelta
df_dynamic['DateTime'] = pd.to_timedelta(df_dynamic['DayNum'] - 1, unit='D') + reference_date

# Create separate columns for date and time
df_dynamic['Date'] = df_dynamic['DateTime'].dt.date

# Create a time column with proper formatting
df_dynamic['Time'] = df_dynamic['DateTime'].dt.time
```
