```python
import os
from google.cloud import discoveryengine
from google.api_core.client_options import ClientOptions
from dotenv import load_dotenv

# 1. Load env
load_dotenv()
PROJECT_ID = os.getenv("PROJECT_ID")
LOCATION = os.getenv("LOCATION")
DATA_STORE_ID = os.getenv("DATA_STORE_ID")

print(f"üîç Checking: Project={PROJECT_ID} | Location={LOCATION} | ID={DATA_STORE_ID}")

# 2. Configure Client
try:
    # Handle the API endpoint (Global vs Regional)
    client_options = (
        ClientOptions(api_endpoint=f"{LOCATION}-discoveryengine.googleapis.com")
        if LOCATION != "global"
        else None
    )

    client = discoveryengine.DataStoreServiceClient(client_options=client_options)
    
    # --- THE FIX: Manually build the path string ---
    # This avoids the "unexpected keyword argument" error
    name = f"projects/{PROJECT_ID}/locations/{LOCATION}/collections/default_collection/dataStores/{DATA_STORE_ID}"
    
    print(f"Attempting to fetch: {name}")
    
    # 3. Try to get the data store details
    response = client.get_data_store(name=name)
    print("\n‚úÖ SUCCESS! Found Data Store:", response.display_name)
    print("Your credentials and IDs are correct.")

except Exception as e:
    print("\n‚ùå FAILED. Here is the error:")
    print(e)
    print("-" * 20)
    print("üí° TIP: If the error says '404' or 'Resource Not Found', switch LOCATION to 'global' in your .env file.")
```
