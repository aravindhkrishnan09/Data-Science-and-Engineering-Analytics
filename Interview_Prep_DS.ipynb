```python
import pandas as pd
import matplotlib.pyplot as plt

a = {'name':['A','B','C'],'age':['25','35','45']}
df = pd.DataFrame(a)
df['age'] = pd.to_numeric(df['age'])

# Create a figure with 3 subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True) # sharey ensures y-axis scale is same

# Iterate through each person and create a subplot
for i, name in enumerate(df['name']):
    ax = axes[i] # Get the current subplot axis
    person_data = df[df['name'] == name]

    ax.bar(person_data['name'], person_data['age'], color=['red', 'green', 'blue'][i])

    # Set y-axis to start from 0
    ax.set_ylim(bottom=0)

    # Add a title for each subplot
    ax.set_title(f'{name}\'s Age Distribution')

    # Add axis labels
    ax.set_xlabel('Name')
    ax.set_ylabel('Age')

    # Display legend for each subplot (optional, can be removed if redundant)
    ax.legend([name])

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler

df =  pd.DataFrame({'x': np.arange(100,1100,100),'y': 2*np.arange(100,1100,100)})
print(df)

x = df[['x']]
y = df[['y']]

x_train, x_test, y_train, y_test = train_test_split(x , y, test_size=0.5, random_state = 42)

model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
print(f"Mean Squared Error:{mse:.4f}")
print("Mean Absolute Error:",mae)
print("R-squared:",r2)

plt.plot(x,y,color='blue',marker = 'o')
plt.plot(x_test,y_pred,color='red',marker = '*')
plt.show()
```
