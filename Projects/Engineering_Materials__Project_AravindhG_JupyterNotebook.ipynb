# Identify the column containing outliers

```python
def identify_outliers(df):

    numeric_columns = df.select_dtypes(include=['int64','float64']).columns

    outlier_columns = []
    for col in numeric_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        if ((df[col] < lower_bound) | (df[col] > upper_bound)).any():
            outlier_columns.append(col)

    print("Columns containing outliers:", outlier_columns)

    print("Number of outliers in each column:")
    for col in outlier_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        print(f"{col}: {outliers_count} outliers")

    return outlier_columns

# Call the function
outlier_columns = identify_outliers(df1)
```

# Plot the columns containing outliers

```python
# plot the columns containing outliers
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_outliers(df, columns):
    for col in columns:
        plt.figure(figsize=(5, 2))
        sns.boxplot(x=df[col], color='lightblue', flierprops=dict(markerfacecolor='red', marker='o'))
        
        # Calculate statistics
        mean_val = df[col].mean()
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Add text labels for Q1, Q3, mean, lower bound, and upper bound
        stats_text = (f"Mean: {mean_val:.2f}\n"
                      f"25th Percentile (Q1): {Q1:.2f}\n"
                      f"75th Percentile (Q3): {Q3:.2f}\n"
                      f"Lower Bound: {lower_bound:.2f}\n"
                      f"Upper Bound: {upper_bound:.2f}")
        plt.gcf().text(0.85, 0.83, stats_text, fontsize=8, color='black', 
                       ha='right', va='top', bbox=dict(facecolor='white', alpha=0.5, edgecolor='gray'))
        
        plt.title(f'Boxplot of {col}')
        plt.xlabel(col)
        plt.show()

plot_outliers(df1, outlier_columns)

# remove file if already exists and  export the df1 to a CSV file
import os
if os.path.exists('Data_cleaned.csv'):
    os.remove('Data_cleaned.csv')
df1.to_csv('Data_cleaned.csv', index=False)
```

# Task 10: Material Ranking by Multi-Criteria Score

```python
df_score = df_merged_subset.copy()

# Manual min-max normalization for positive indicators
df_score['Su_norm'] = (df_score['Su'] - df_score['Su'].min()) / (df_score['Su'].max() - df_score['Su'].min())
df_score['A5_norm'] = (df_score['A5'] - df_score['A5'].min()) / (df_score['A5'].max() - df_score['A5'].min())

# Reverse normalization for negative indicator (Ro - lower is better)
df_score['Ro_norm'] = 1 - ((df_score['Ro'] - df_score['Ro'].min()) / (df_score['Ro'].max() - df_score['Ro'].min()))

# Neutralization score for pH (closer to 7 is better)
df_score['pH_score'] = 1 - abs(df_score['pH'] - 7) / 7

# Penalty for extreme Poisson's ratio (ideal ~0.3)
df_score['mu_penalty'] = abs(df_score['mu'] - 0.3)

# Compute weighted material score
df_score['Material_Score'] = (
    0.35 * df_score['Su_norm'] +
    0.25 * df_score['A5_norm'] +
    0.15 * df_score['Ro_norm'] +
    0.15 * df_score['pH_score'] -
    0.10 * df_score['mu_penalty']
)

# Rank materials
df_score['Rank'] = df_score['Material_Score'].rank(method='dense',ascending=False)

# Show top 5
print(df_score[['Material_Alias', 'Material_Score', 'Rank']].sort_values(by='Rank').head(10))
```

# Task 12: Material Descriptor Analysis

```python
import pandas as pd
import numpy as np

# Sample keywords to search
keywords = ["ductile", "hardening", "corrosion-resistant"]

# Prepare a dictionary to store results
descriptor_stats = {}

# Make sure Desc is a string and lowercase
df_merged_subset['Desc'] = df_merged_subset['Desc'].astype(str).str.lower()

for word in keywords:
    mask = df_merged_subset['Desc'].str.contains(word)
    count = mask.sum()
    avg_properties = df_merged_subset[mask][['Su', 'Sy', 'A5', 'E', 'G', 'Ro']].mean(numeric_only=True)
    descriptor_stats[word] = {
        "count": count,
        "avg_properties": avg_properties
    }

# Display summary
for key, value in descriptor_stats.items():
    print(f"\nKeyword: {key.capitalize()}")
    print(f"Matching materials: {value['count']}")
    print("Average Mechanical Properties:")
    print(value['avg_properties'].round(2))

# Plotting
fig, axes = plt.subplots(len(keywords), 1, figsize=(10, 12))
for i, word in enumerate(keywords):
    avg_vals = descriptor_stats[word]['avg_properties']
    bars = axes[i].bar(avg_vals.index, avg_vals.values, color='skyblue')
    axes[i].set_title(f'Average Properties for Keyword: {word.capitalize()} (n={descriptor_stats[word]["count"]})')
    axes[i].set_xlabel('Properties')
    axes[i].set_ylabel('Value')
    axes[i].tick_params(axis='x', rotation=45)
    
    # Add value labels to each bar
    for bar in bars:
        height = bar.get_height()
        axes[i].text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
