# Broadcasting

```python
#scalar value on an one dimentional array

arr = np.array([1,2,3,4,5])
print(arr+2)
print(arr-2)
print(arr*2)
print(arr/2)
print(arr//2)
print(arr%2)
print(arr**2)
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2,3)
B = np.array([10, 20])  # Shape (2,)
print(B)
#result = A + B
result = A + B.reshape(2,1)
print(A)
print(B.reshape(2,1))
print(result)
```

#  Reshaping Arrays

```python
arr = np.arange(1,7)
print(arr)
arr = arr.reshape(2,3)
print(arr)
arr = arr.reshape(3,2)
print(arr)
arr = arr.flatten()
print(arr)
```

# Transposing Arrays

```python
arr = np.arange(1,7)
print(arr)
arr = arr.reshape(2,3)
print(arr)
arr = arr.transpose()
print(arr)
```

# Concatenating Arrays

```python
arr_a = np.arange(1,4)
arr_b = np.arange(4,7)
print(arr_a)
print(arr_b)
arr_c = np.concatenate((arr_a,arr_b))
arr_d = np.concatenate((arr_b,arr_a))
print(arr_c,arr_d)
```

# Splitting Arrays

```python
arr_a = np.arange(1,7)
print(arr_a)
arr_b = np.split(arr_a,2)
print(arr_b)
arr_c = np.array(arr_b)
print(arr_c)
arr_d = np.array(arr_b).tolist()
print(arr_d)
```

# Mathematical Functions

```python
arr_a = np.arange(1,4)
arr_b = np.arange(4,7)
print(arr_a)
print(arr_b)
arr_add = np.add(arr_a,arr_b)
arr_sub = np.subtract(arr_a,arr_b)
arr_mul = np.multiply(arr_a,arr_b)
arr_div = np.divide(arr_a,arr_b)
arr_mod = np.mod(arr_a,arr_b)
arr_pow = np.power(arr_a,arr_b)
print(arr_add)
print(arr_sub)
print(arr_mul)
print(arr_div)
print(arr_mod)
print(arr_pow)
```

# Correlation and Covariance

```python
import numpy as np
arr_a = np.random.randint(1,10,6)
arr_b = np.random.randint(1,10,6)
print(arr_a)
print(arr_b)
print(np.corrcoef(arr_a,arr_b))
print(np.cov(arr_a,arr_b))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data: Sales vs Profit (Hypothetical)
sales = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
profit = np.array([10, 40, 50, 80, 100, 150, 180, 200, 230, 260])  # Positive correlation

# Compute correlation coefficient
corr_matrix = np.corrcoef(sales, profit)
correlation_coefficient = corr_matrix[0, 1]

# Plot the data
plt.figure(figsize=(7,5))
plt.scatter(sales, profit, color='blue', label='Sales vs Profit')
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.title(f"Correlation between Sales and Profit\nCorrelation Coefficient: {correlation_coefficient:.2f}")
plt.grid(True)
plt.legend()
plt.show()

# Print correlation coefficient
print(f"Correlation Coefficient: {correlation_coefficient:.2f}")
```
