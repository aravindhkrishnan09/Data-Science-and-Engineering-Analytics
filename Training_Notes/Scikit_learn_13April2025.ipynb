# Statistics and Probability

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm, poisson

# Binomial PMF
n, p = 10, 0.5
x_binom = np.arange(0, n + 1)
y_binom = binom.pmf(x_binom, n, p)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.bar(x_binom, y_binom, color='blue', alpha=0.7)
plt.title('Binomial PMF (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

# Normal PDF
x_norm = np.linspace(-4, 4, 100)
y_norm = norm.pdf(x_norm, 0, 1)
plt.subplot(1, 3, 2)
plt.plot(x_norm, y_norm, color='green')
plt.title('Normal PDF (mean=0, std=1)')
plt.xlabel('x')
plt.ylabel('Density')

# Poisson PMF
lambda_poisson = 2
x_poisson = np.arange(0, 10)
y_poisson = poisson.pmf(x_poisson, lambda_poisson)
plt.subplot(1, 3, 3)
plt.bar(x_poisson, y_poisson, color='red', alpha=0.7)
plt.title('Poisson PMF (Î»=2)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 175, 10
x = np.linspace(140, 210, 500)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y)
plt.title("Distribution of Male Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.axvline(185, color='r', linestyle='--', label='185 cm (Z=1)')
plt.legend()
plt.grid()
plt.show()
```
